```python
import numpy as np
import sys

print("="*29)
print("| Program Linier Regression |")
print("="*29)
n = int(input("Masukan Nilai N = "))
A = B = C = D = E = r = b = a = Y = 0

x = np.zeros((n))
y = np.zeros((n))

for i in range(n):
  x[i]=float(input("X["+str(i+1)+"] : "))
  y[i]=float(input("Y["+str(i+1)+"] : "))
  print("="*10)
  A = A + x[i] * y[i]
  B = B + x[i]
  C = C + y[i]
  D = D + x[i]**2
  E = E + y[i]**2


print("\n")
print("="*29)
print("Operasi perhitungannya")
print("="*29)
print("")


r = r + (n * A - B *C)/((n*D-B**2) * (n*E-C**2))**0.5
print("r = (",n,"*",A,"-",B,"*",C,")/((",n,"*",D,"-",B,"^2 ) * (",n,"*",E,"-",C,"^2 )) ^ 0.5")
print("r = ", r)
print("")
b = (n*A-B*C)/(n*D-B**2)
print("b = (",n,"*",A,"-",B,"*",C,")/(",n,"*",D,"-",B,"^2)")
print("b = ",b)
print("")
a = (C*D-B*A)/(n*D-B**2)
print("a = (",C,"*",D,"-",B,"*",A,")/(",n,"*",D,"-",B,"^2)")
print("a = ",a)
print("")
Y = a + b
print("Y = ",a,"+",b)
print("Y = ",Y)
```
