```python
import numpy as np
import sys

n = int(input("Masukan Jumlah Titik : "))

a = np.zeros((n,n+1))
b = np.zeros((n))
for i in range(n):
  for j in range(n+1):
    a[i][j] = float(input('a['+str(i)+']['+str(j)+']='))

#Penerapan Eliminasi Gauss-Jordan
for i in range(n):
  if (a[i][i] == 0.0):
    sys.exit('Pembagian dengan Nol!')
  for j in range(n):
    if (i!=j):
      ratio = a[j][i]/a[i][i]
      for k in range(n+1):
        a[j][k] = a[j][k] - ratio*a[i][k]
print('Tampilan hasil pengolahan sebelum menjadi matriks Identitas')

for i in range(n):
  for j in range(n+1):
        print(round(a[i][j], 3)," ",end='')
  print("\n")
print('Tampilan hasil Pengolahan Setelah Menjadi Matriks Identitas')

for i in range(n):
  nilai_diagonal = a[i][i]
  for j in range(n+1):
    a[i][j] = a[i][j]/nilai_diagonal
    print(round(a[i][j], 3)," ",end='')
    b[i]=round(a[i][n],3)
  print("\n")

print("tampilan hasil koefisien dari persamaan polinom")
#hasil nilai a setiap baris
for i in range(n):
  print("a",i,":",b[i])

y = 0
x=float(input("Nilai X = "))
for i in range(n):
  y += b[i] * x **i

print("Nilai Y = ", y)
```
